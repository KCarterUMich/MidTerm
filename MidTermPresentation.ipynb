```python
# prompt: rename the column 'Year ' to remove the space

import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023 = pd.read_csv('/content/drive/MyDrive/MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.csv')

# Rename the column 'Year ' to 'Year'
MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023 = MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.rename(columns={'Year ': 'Year'})

# Verify the change
print(MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.columns)
```

```python
# prompt: Print a bar graph comparing 2019, 2020, 2021, 2022 Total Sales New and Total Sales Used

import matplotlib.pyplot as plt

# Group the data by year and sum the sales for 'Total Sales New' and 'Total Sales Used'
yearly_sales = MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.groupby('Year')[['Total Sales New', 'Total Sales Used']].sum()

# Select the years 2019, 2020, 2021, and 2022
years_to_plot = [2019, 2020, 2021, 2022]
sales_to_plot = yearly_sales.loc[years_to_plot]

# Create the bar graph
sales_to_plot.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of New and Used Vehicle Sales (2019-2022)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Sales Type')
plt.grid(True)
plt.show()
```

```python
# prompt: Change the number on the y axis to not be in scientific notation

import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_file.csv' with the path to your CSV file
MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023 = pd.read_csv('/content/drive/MyDrive/MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.csv')

# Rename the column 'Year ' to 'Year'
MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023 = MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.rename(columns={'Year ': 'Year'})

# Group the data by year and sum the sales for 'Total Sales New' and 'Total Sales Used'
yearly_sales = MVA_Vehicle_Sales_Counts_by_Month_for_Calendar_Year_2002_through_December_2023.groupby('Year')[['Total Sales New', 'Total Sales Used']].sum()

# Select the years 2019, 2020, 2021, and 2022
years_to_plot = [2019, 2020, 2021, 2022]
sales_to_plot = yearly_sales.loc[years_to_plot]

# Create the bar graph
ax = sales_to_plot.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of New and Used Vehicle Sales (2019-2022)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Sales Type')
plt.grid(True)

# Format y-axis to display numbers without scientific notation
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

plt.show()
```
